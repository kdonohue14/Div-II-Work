## Clickback Email Health


```python
clickback_csv = [xml for xml in glob.iglob('zoominfo_permissionpass*.csv')]
clmns = ['Import_Date','Email_Health','Bounce_Code','Verification_Result','ZoomIndividualID','Email_Address']
```

```python
Email_Health_Trends = pd.DataFrame()

for i in clickback_csv:
    Email_Health_Trends = Email_Health_Trends.append(pd.read_csv(i, usecols=clmns))
```

Here we wanted the counts of email_health values in each files broken out by date.

```python
Trend_Analysis = Email_Health_Trends.groupby('Import_Date')['Email_Health'].value_counts().unstack()
```

```python
Trend_Analysis.head()
```

Adds new columns with percentages of row sum 

```python
Trend_Analysis['Totals'] = Trend_Analysis.sum(axis=1)
Trend_Analysis2 = Trend_Analysis.apply(lambda x : x / Trend_Analysis['Totals'] * 100)
Trend_Analysis = pd.merge(Trend_Analysis, Trend_Analysis2, how='left', left_index=True, right_index=True, suffixes = ['',' Percentage'])
Trend_Analysis
```

```python
writer = pd.ExcelWriter('Clickback_Trend_Analysis_new.xlsx')
Trend_Analysis.to_excel(writer, 'Trend_Analysis')
writer.save() 
```
